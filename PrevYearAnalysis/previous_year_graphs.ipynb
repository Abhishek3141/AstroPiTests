```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('previous_year_data.csv')
print(df.head())
```

```python
long= df["Postion longitude"]
lat = df["Position latitude"]

plt.plot(long,lat,'o', ms='0.5')
```

```python
Bx= df["Magnetic field strenght X"]
By= df["Magnetic field strenght Y"]
Bz= df["Magnetic field strenght Z"]

B = (Bx**2 + By**2+Bz**2)**0.5
```

```python

plt.plot(lat, B, 'o', ms = '1', c='r')
```
